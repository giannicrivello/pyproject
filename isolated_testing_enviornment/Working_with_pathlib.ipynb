```python
#makes a single or nested directory
import pathlib

def make_dir(*args):
    p = pathlib.Path(*args)
    try:
        p.mkdir(parents=True)
    except FileExistsError:
        print("making more of the same dirs you fucjking animal")
```

```python
make_dir("Parent1/Child1/child2")
```

```python
#walking a directory tree
import os
for dirpath, dirnames, files in os.walk('.', topdown=False):
    print(f'Found directory: {dirpath}')
    for file_name in files:
        print(file_name)
```
